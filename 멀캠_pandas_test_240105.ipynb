```python
tips.loc[:4,['total_bill', 'tip', 'day']]
```

## (6) 조건에 따른 행 추출
- 'day'가 'Sunday'인 모든 행을 추출하고 'total_bill'과 'tip' 열만 표시하세요.

```python
data = tips.loc[tips['day']=='Sun',:]
data.loc[:,['total_bill', 'tip']]
```

## (7) 다중 조건에 따른 행 추출
- loc을 사용하여 'time'이 'Dinner'이고 'tip'이 5달러 이상인 모든 행을 선택하고 'total_bill', 'sex', 'smoker' 열만 표시하세요.

```python
data2 = tips.loc[(tips['tip']>=5)&(tips['time']=='Dinner')]
data2.loc[:,[ 'total_bill', 'sex', 'smoker' ]].reset_index(drop = True)
```
